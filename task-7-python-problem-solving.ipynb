# Problem 1: Smart Light Controller

Problem Statement-
Create a class to control a smart light that can be turned ON or OFF and display its status.

üåç Real-Time Use-
Smart home automation systems

üì• Sample Input-
Light Name: Bedroom Light,
Action: ON

üì§ Expected Output-
Bedroom Light is ON


```python
class Smartlight:
   def __init__(self, lightname, action):
     self.lightname = lightname
     self.action = action

   def controller(self, action):
    if action == "ON":
       print(f"{self.lightname} is ON")
    elif action == "OFF":
       print(f"{self.lightname} is OFF")
    else:
       print("Invalid action")

lightname = input("Light Name: ")
action = input("Action: ").upper()

smartlight = Smartlight(lightname, action)
smartlight.controller(action)

    
```

# Problem 2: Employee ID Card System

Problem Statement-
Create a class that stores employee name, ID, and department and displays ID card details.

Real-Time Use-
Corporate HR systems

üì• Sample Input-
Employee Name: Rahul,
Employee ID: EMP102,
Department: IT

üì§ Expected Output-
Employee ID Card,
Name: Rahul,
ID: EMP102,
Department: IT


```python
class employee:
    def __init__(self, name, ID, Dept):
      self.name = name
      self.ID = ID 
      self.Dept = Dept

    def display(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.ID}")
        print(f"Dept: {self.Dept}")

name = input("Employee Name: ")
ID = input("Employee ID: ")
Dept = input("Employee Department: ")

employee = employee(name, ID, Dept)
employee.display()
        
```

# Problem 3: Mobile Contact Record

Problem Statement-
Create a class to store contact name and phone number and display contact information.

Real-Time Use-
Mobile contact applications

üì• Sample Input-
Contact Name: Anita,
Phone Number: 9876543210

üì§ Expected Output-
Contact Saved,
Name: Anita,
Phone: 9876543210





```python
class contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display(self):
        print("Contact Saved")
        print(f"Contact Name: {self.name}")
        print(f"Phone number: {self.phone_number}")

name = input("Contact name: ")
phone_number = int(input("Phone number: "))

contact = contact(name, phone_number)
contact.display()
```

# Problem 4: Product Price Tag Generator

Problem Statement-
Create a class that stores product name and price and prints a formatted price tag.

 Real-Time Use-
Retail billing systems

üì• Sample Input-
Product Name: Headphones,
Price: 2499

üì§ Expected Output-
Product: Headphones,
Price: ‚Çπ2499


```python
class product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    def price_tag(self):
        print("OUTPUT")
        print(f"Product: {self.name}")
        print(f"Price: {self.price}")

name = input("Product Name: ")
price = int(input("Price: "))

product = product(name, price)
product.price_tag()
        
```

# Problem 5: Movie Rating Display System

Problem Statement-
Create a class that stores movie name and rating and displays movie details.

Real-Time Use-
Streaming platforms

üì• Sample Input-
Movie Name: Inception,
Rating: 4.8

üì§ Expected Output-
Movie: Inception,
Rating: 4.8 / 5




```python
class movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
        
    def display(self):
        print("MOVIE DETAILS:")
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating}")

name = input("Movie Name: ")
rating = float(input("Rating: "))

movie = movie(name, rating)
movie.display()
```
