# Problem 1: Mobile Recharge Validation System

Problem Statement-
Create a function to validate a mobile recharge request.

Rules:
Recharge amount must be ≥ ₹50

Valid plans: 199, 299, 399, 599

User should be allowed to retry if invalid

Real-Time Use-
Telecom recharge apps,
Payment gateways

Hint-
Use function for validation,
Use while loop for retry

```python
def validate_recharge():
    valid_plans= [199, 299, 399, 599]
    while True:
        amt = int(input("Enter Recharge Amount:"))
        if amt < 50:
            print("Invalid amount. Minimum recharge is Rs.50")
        elif amt not in valid_plans:
            print("Invalid plan. Choose from this- 199, 299, 399, 599")
        else:
            print("Recharge successful")
            break

validate_recharge()
    
```

# Problem 2: Inventory Reorder Alert System

Problem Statement-
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15,
Print Stock OK otherwise

Real-Time Use-
Warehouse systems,
Retail management


Hint-
Use function,
Use for loop on dictionary





```python
def inventory_check(products):
    for product, stock in products.items():
        if stock<15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

products={
    "Laptop":20,
    "Mouse":10,
    "Keyboard":25,
    "Monitor":8
}
inventory_check(products)
```

# Problem 3: Student Result Processing System

Problem Statement-
Create a function that:
Accepts a list of student marks

Calculates average marks

Prints Pass if average ≥ 50 else Fail

Real-Time Use-
Online exam systems,
Evaluation portals

Hint-
Use loop to calculate total,
Return result from function




```python
def result(marks_list):
    total=0
    for mark in marks_list:
        total+=mark
    average=total/len(marks_list)

    print("Average Marks:",average)

    if average>=50:
        return "Pass"
    else:
        return "Fail"
    
user_input=(input("Enter the marks(space seperated): "))
marks=list(map(int, user_input.split()))
final_result=result(marks)
print("Result: ",final_result)
```

# Problem 4: Cab Fare Estimator with Retry Option

Problem Statement-
Create a cab fare calculator:
Base fare = ₹50,
₹12 per km,
Peak hour → 25% extra

Allow retry using while loop

Real-Time Use-
Ride-hailing apps,
Transport systems

Hint-
Function for fare calculation,
While loop for retry












```python
def fare_calculation():
    while True:
        try:
            distance=float(input("Enter the distance traveled in km: "))
            peak=input("Is it peak hour? (yes/no): ").lower()

            fare=50+(12*distance)

            if peak=="yes":
                fare+=fare*0.25

            print(f"Total Fare: ₹{fare}")
            break
        except:
            print("Invalid input. Please try again.")

fare_calculation()
```

# Problem 5: Employee Attendance Eligibility Checker

Problem Statement-
Create a function that:
Accepts attendance list ("P", "A")

Calculates attendance percentage

Returns Eligible if ≥ 75% else Not Eligible

Real-Time Use-
HR payroll systems,
Internship tracking

Hint-
Use loop to count present days,
Use return statement





```python
def check_attendance(attendance):
    present=0
    for day in attendance:
        if day=="P":
            present+=1
        
    percentage=(present/len(attendance))*100
    print("Attendance Percentage:",percentage)

    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance=["P","P","A","P","P"]
status=check_attendance(attendance)
print("Eligibility Status: ",status)
```

# Problem 6: Password Strength Checker

Problem Statement-
Create a function to check password strength.
Rules:
Minimum length = 8

Must contain at least one digit

Must contain at least one special character (@#$)

Real-Time Use-
Login systems,
Security validation

Hint-
Use loop over string,
Use conditions inside function




```python
def check_password(password):
    special_chars="@#$"

    has_digit=False
    has_special=False
    for char in password:
        if char.isdigit():
            has_digit=True
        if char in special_chars:
            has_special=True

    if len(password)>=8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "weak Password"

password=input("Enter Password: ")
print(check_password(password))
```
